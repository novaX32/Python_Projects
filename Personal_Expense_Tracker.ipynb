```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_name="expenses.csv"
def  add_expences():
    date=input("Enter the date (YYYY-MM-DD): ")
    category=input("Enter the category of the expences eg (food,travel,entertainment,girlfriend)")
    amount=float(input("Enter the amount: "))
    desc=input("Enter the description: ")
    expence=pd.DataFrame([[date,category,amount,desc]],columns=['Date','Category','Amount','Description'])
    try:
        df=pd.read_csv(file_name)
        df=pd.concat([df,expence],ignore_index=True)
    except FileNotFoundError:
        df=expence
    except pd.errors.EmptyDataError:
        df=expence
    df.to_csv(file_name,index=False)
    print("Successfully Added Data")

def view_expences():
    try:
        df=pd.read_csv(file_name)
        print("Your Expences\n")
        print(df)
    except FileNotFoundError:
        Print("No Exception Data Yet\n")

def analyze():
    try:
        df=pd.read_csv(file_name)
        cat_total=df.groupby('Category')['Amount'].sum()
        
        plt.figure(figsize=(10,5))
        cat_total.plot(kind='bar',color='skyblue')
        plt.title("Total Amount spend per category")
        plt.ylabel("Amount")
        plt.show()

        plt.figure(figsize=(10,5))
        cat_total.plot(kind='pie',autopct="%1.1f%%", colormap="viridis")
        plt.title("Amount Spend Destribution")
        plt.ylabel("")
        plt.show()
        
    except FileNotFoundError:
        print("No Expence Data Yet")
    
```
